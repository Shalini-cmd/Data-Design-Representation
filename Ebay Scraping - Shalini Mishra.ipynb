```python
import numpy as np

def scrape_store(path,sponsor_flag):
    scrape_df=pd.DataFrame(columns=['seller_name', 'seller_score', 'item_price','units_sold','best_offer',
                     'item_title','returns_allowed','shipping_price','item_condition','sponsored'])
    #Iterating through files in a directory
    for file in os.listdir(path):
        #open each file in read bytes mode
        f=open(file, "rb")
        filetext=f.read()
        #parse it through Beautiful soup 
        data = BeautifulSoup(filetext, 'html.parser')

        #Get Seller-name
        seller_name='No info'if data.find('span',class_='mbg-nw') is None else data.find('span',class_='mbg-nw').string

        #store seller score in integer format
        seller_score=np.nan if data.select('span.mbg-l > a') is None else int(data.select('span.mbg-l > a')[0].string)

        #Item Price
        item_price= np.nan if data.find('span',class_='notranslate')is None else data.find('span',class_='notranslate').string
        #,itemprop='price'
        #if NoneType object is found then store as nan(value unavailable)
        units_sold= np.nan if data.find('a',class_='vi-txt-underline') is None else data.find('a',class_='vi-txt-underline').string

        #if we get NoneType object, best offer doesn't exist for the item
        best_offer= True if data.find('div',class_=re.compile('.*boLable')) is not None else False

        #Item title
        title1=data.find('h1',class_='it-ttl',id='itemTitle').get_text()
        #removing unecessary part of the string
        item_title=re.sub('.*\xa0', '', title1)

        #Return Details
        returns_allowed= None if data.find('span',id='vi-ret-accrd-txt') is None else data.find('span',id='vi-ret-accrd-txt').string

        #Shipping Price
        #findinf shipping summary then looking for next as its dynamic across product pages
        shipping= data.find('span',id='shSummary')
        shipping_price=shipping.find_next().get_text()

        #Item Condition
        item_condition=data.find('div',id='vi-itm-cond',itemprop='itemCondition').string

        #Close the file
        f.close()

        #Append to the dataframe
        scrape_df = scrape_df.append({'seller_name': seller_name, 'seller_score': seller_score, 'item_price':item_price,
                                 'units_sold':units_sold,'best_offer':best_offer,'item_title':item_title,
                                 'returns_allowed':returns_allowed,'shipping_price':shipping_price,'item_condition':item_condition,
                                'sponsored':sponsor_flag}, ignore_index=True)
    return(scrape_df)
```

#### Created a Function that scrapes the files in the directory and adds to the MySQL table created above
Passing Parameters: path to the directory, value for sponsor(1:sponsored or 0:non-sponsored), cursor(mySQL cursor)


```python
def scrape_storedb(path,sponsor_flag,cursor):
    #for each file in the directory getting the data  and inserting into the sql table eBay_items
    for file in os.listdir(path):
        #open each file in read bytes mode
        f=open(file, "rb")
        filetext=f.read()
        #parse it through Beautiful soup 
        data = BeautifulSoup(filetext, 'html.parser')

        #Get Seller-name
        seller_name=None if data.find('span',class_='mbg-nw') is None else data.find('span',class_='mbg-nw').string

        #store seller score in integer format
        seller_score=np.nan if data.select('span.mbg-l > a') is None else int(data.select('span.mbg-l > a')[0].string)

        #ITEM PRICE
        item_price= np.nan if data.find('span',class_='notranslate')is None else data.find('span',class_='notranslate').string
        #,itemprop='price'

        item_price = re.sub('\n+|\t+|,|\$|\/ea','',item_price)
        #split the column into currency type  and amount
        item_c= None if (re.search('[A-Z]+',item_price)) is None else (re.search('[A-Z]+',item_price)).group(0)
        item_price=(re.search('[0-9]+.[0-9]+',item_price)).group(0)
        
        if item_c=='US':
            currency=1
        elif item_c=='AU':
            currency=0.67
        elif item_c=='C':
            currency=0.75
        elif item_c=='GBP':
            currency=1.25
        else:currency=1 #Cases where no country is mentioned, default as US

        #Changing the data type
        item_price=float(item_price)*currency*100
        item_price=int(item_price)

        #UNITS SOLD
        #if NoneType object is found then store as nan(value unavailable)
        units_sold= None if data.find('a',class_='vi-txt-underline') is None else data.find('a',class_='vi-txt-underline').string
        #Regex for sold substring from the quantities
        units_sold= units_sold if units_sold is None else re.sub(' sold','',units_sold)

        #BEST OFFER AVAILABLE
        #if we get NoneType object, best offer doesn't exist for the item
        best_offer= True if data.find('div',class_=re.compile('.*boLable')) is not None else False
        #Best Offers as 0/1
        best_offer= int(1 if best_offer==True else 0)

        #Item title
        title1=data.find('h1',class_='it-ttl',id='itemTitle').get_text()
        #removing unecessary part of the string
        item_title=re.sub('.*\xa0', '', title1)

        #RETURNS DETAILS
        returns_allowed= None if data.find('span',id='vi-ret-accrd-txt') is None else data.find('span',id='vi-ret-accrd-txt').string
        #Returns Allowed-0/1
        returns_allowed= int (0 if returns_allowed =='Seller does not accept returns' else 1)

        #SHIPPING PRICE
        #finding shipping summary then looking for next as its dynamic across product pages
        shipping= data.find('span',id='shSummary')
        shipping_price=shipping.find_next().get_text()

        #converting into integers
        shipping_price = re.sub('\n+|\t+','',shipping_price)
        shipping_price = re.sub('(AU|GBP|C)( ){1,2}\${0,1}[0-9]+.[0-9]+\(approx.( ){0,2}US|\)|\$','',shipping_price)

        #if the shipping is variable 1 else 0
        shipping_zip= 1 if shipping_price in('Local pick-up offered.','Calculate','Free Local Pickup') else 0

        #All free shipping have the value 0
        def shipping(x):
            if x =='FREE': 
                return 0
            elif x =='FAST \'N FREE': 
                return 0
            elif x=='Calculate': 
                return 0
            elif x=='Local pick-up offered.': 
                return 0
            elif x=='Free Local Pickup':
                return 0
            else: return  x

        #Changing data-type
        shipping_price=shipping(shipping_price)
        shipping_price= float(shipping_price)
        #Converting to INT
        shipping_price= int(shipping_price*100)

        #Item Condition
        item_condition= None if data.find('div',id='vi-itm-cond',itemprop='itemCondition') is None else data.find('div',id='vi-itm-cond',itemprop='itemCondition').string

        #sponsor flag
        sponsored=sponsor_flag

        #Close the file
        f.close()

        #inserting into the table omdb
        sql_query = "INSERT INTO eBay_items(item_title, item_price, seller_name,seller_score,units_sold,returns_allowed,best_offer,shipping_price,shipping_zip,item_condition,sponsored) VALUES(%s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s)"
        record=(item_title, item_price, seller_name,seller_score,
                units_sold,returns_allowed,best_offer,shipping_price,
                shipping_zip,item_condition,sponsored)
        cursor.execute(sql_query,record)
        #print("inserted")
        #print("\n")

    
```

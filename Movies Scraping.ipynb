<h5>(a) Write a program in Python or Java that uses the OMDb API to lookup/search _your_ top 10 choices of movie titles and extract their imdbIDs. Print the movie titles along with their imdbID to the screen. (E.g., search for "Dangal" or "Wolf Warrior 2" and record the imdbID that corresponds to the exact movie you have in mind).</h5><br>


```python
#importing necessary libraries
import requests as r
import json as js
top_movies=['Life of pi','Her','Orphan','The GodFather','Moonlight','Get out',
            'Age Of Adaline','Shawshank Redemption','Seven','The Curious Case Of Benjamin Button']

#creating an empty list to store movie IDs
mov_ID=[]
for i in range(len(top_movies)):
        url="http://omdbapi.com/?t="+top_movies[i]+"&apikey=ce1b31ce&type=movie"
        #Request to the server
        response = r.get(url)
        result=js.loads(response.text)
        print("Title:",result['Title'])
        print("IMDB ID:",result['imdbID'])
        mov_ID.append(result['imdbID']) 
        print("\n")
```

<h5>(c) "Recycle" + augment your code from (2.d) to insert the following information of _your_ top 10 movie titles into a new SQL table named "omdb":<br>title, year, genre, director, imdb_rating, rotten_tomatoes, metacritic, plot, box_office.<br>
Please store all ratings (imdb_rating, rotten_tomatoes, metacritic) all as full integers representing percentages, i.e., 8.5/10 = 85, 80% = 80<br></h5><br>


```python
#Connecting to MySQL using Python connector
import mysql.connector

#establishing the connection
#entering the database names as ucdavis is itsn't there it gets created automatically
conn = mysql.connector.connect(
   user='root', password='', host='localhost', database='ucdavis'
)

#Creating a cursor object using the cursor() method
cursor = conn.cursor()

#Dropping OMDB_test table if already exists.
cursor.execute("DROP TABLE IF EXISTS ucdavis.OMDB")

#Creating table as per requirement
sql ='''CREATE TABLE OMDB(
   imdb_id VARCHAR(20) PRIMARY KEY NOT NULL ,
   title VARCHAR(600),
   year INT(4),
   genre VARCHAR(255), 
   director VARCHAR(255), 
   imdb_rating INT(2), 
   rotten_tomatoes INT(2), 
   metacritic INT(2), 
   plot VARCHAR(1000), 
   box_office VARCHAR(20)
)'''
cursor.execute(sql)

#Checking where the cursor is at
for table in cursor:
    print(table)
    
```

```python
print(cursor)
```

```python
#for each movie in the list mov_ID getting the data from api and inserting into the sql table omdb
for movie in mov_ID:
    url = "http://www.omdbapi.com/?apikey=ce1b31ce&i="+movie+"&type=movie"
    
    #Request to the server to get data
    response = r.get(url)
    my_mov=response.json()
 
    #converting ratings into integers
    imdb_rating=float((my_mov['Ratings'][0]['Value']).split("/")[0])*10
    rottentomatoes_rating=float((my_mov['Ratings'][1]['Value']).split("%")[0])
    metacritic_rating=float((my_mov['Ratings'][2]['Value']).split("/")[0])
    
    #inserting into the table omdb
    sql_query = "INSERT INTO omdb(imdb_id,title,year,genre,director, imdb_rating, rotten_tomatoes, metacritic, plot, box_office) VALUES(%s, %s, %s, %s, %s, %s,%s, %s, %s, %s)"
    record=(movie,my_mov['Title'],my_mov['Year'],my_mov['Genre'],my_mov['Director'],
            imdb_rating,rottentomatoes_rating, metacritic_rating,
            my_mov['Plot'],my_mov['BoxOffice'])
    cursor.execute(sql_query,record)
    print("inserted")
    print("\n")

    
```

```python
conn.commit()

#validating whether records were inserted
cursor.execute("select * from omdb")
records = cursor.fetchall()
print("Total number of rows in omdb is: ", cursor.rowcount)
```
